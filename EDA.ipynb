### Gráficos

Número de fraud (True / False)

```python
count_fraud = df.groupby("fraud_bool").size().reset_index()

print(count_fraud)

sns.barplot(data=count_fraud, x=count_fraud["fraud_bool"], y=count_fraud[0], hue=count_fraud["fraud_bool"])
plt.title("Cantidad de casos por fraud bool")
```

Frecuencia de las edades: la edad que más frecuente en el dataset es 30 años.

```python
edades = df["customer_age"].value_counts().sort_index().reset_index()

sns.barplot(data=edades, x=edades["customer_age"], y=edades["count"])
plt.title("Frecuencia de edades")
```

Credit risk score medio por edad: de media, mayores edades indican un credit risk score mayor, por lo que tienen menos riesgo de impago. Esto puede ser debido a que a mayor edad mayor tendencia a ahorrar y menor tendencia a gastar el dinero compuslivamente. (Suponiendo que a mayor credit risk score --> menor probabilidad de impago)

```python
group_edad_risk_score_medio = df.groupby("customer_age")["credit_risk_score"].mean().reset_index()

sns.barplot(data=group_edad_risk_score_medio, x=group_edad_risk_score_medio["customer_age"], y=group_edad_risk_score_medio["credit_risk_score"])
plt.title("Credit Risk Score medio por edad")
```

Ingresos medios por edad: aquellos que cuentan con mayores ingresos son los de la franja de 40 años. Esto puede deberse a que en estas edades los puestos laborales son más estables además de que se llega al máximo pico de la carrera laboral.

```python
edad_income_medio = df.groupby("customer_age")["income"].mean().reset_index()

sns.barplot(data=edad_income_medio, x=edad_income_medio["customer_age"], y=edad_income_medio["income"])
plt.title("Ingresos medios por edad")
```

% de fraudes por edad: a mayor edad mayor proporción de fraudes, esto puede deberse a que a mayor edad se cuentan con mayor número de préstamos, activos...

```python
total_por_edad = df.groupby("customer_age").size().reset_index()

df_bool_true = df[df["fraud_bool"] == 1]

fraudes = df_bool_true.groupby("customer_age").size().reset_index()

porcentajes = total_por_edad.merge(fraudes, on="customer_age", how="left")

print(porcentajes)

porcentajes["porcentaje_fraude"] = (porcentajes["0_y"] / porcentajes["0_x"]) * 100

count_bool_edad = df_bool_true.groupby("customer_age").size().reset_index()

print(count_bool_edad)

sns.barplot(data=porcentajes, x=porcentajes["customer_age"], y=porcentajes["porcentaje_fraude"], color="red")
plt.title(r"% de fraudes por edad")

```
